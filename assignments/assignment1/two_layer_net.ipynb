## Inline Question 2: 

Now that you have trained a Neural Network classifier, you may find that your testing accuracy is much lower than the training accuracy. In what ways can we decrease this gap? Select all that apply.

1. Train on a larger dataset.
2. Add more hidden units.
3. Increase the regularization strength.
4. None of the above.

$\color{blue}{\textit Your Answer:}$ 1 and 3

$\color{blue}{\textit Your Explanation:}$
If your testing accuracy is much lower than your training accuracy, it indicates that your model is overfitting. Overfitting occurs when the model performs well on the training data but poorly on unseen data because it has learned to memorize the training data rather than generalize to new data. Here are the ways you can decrease this gap:

Train on a larger dataset:
Applicable: Training on a larger dataset can help reduce overfitting. More data helps the model to generalize better by exposing it to a wider variety of examples. This can help the model learn more robust features and improve its performance on the test set.

Add more hidden units:
Not Applicable: Adding more hidden units (increasing the model complexity) can actually worsen overfitting, especially if your model is already complex. More hidden units can lead to the model learning to fit the noise in the training data, which increases the gap between training and testing accuracy.

Increase the regularization strength:
Applicable: Increasing the regularization strength (such as L2 regularization) can help reduce overfitting by penalizing large weights, which discourages the model from fitting the noise in the training data. This helps the model generalize better to the test set.
